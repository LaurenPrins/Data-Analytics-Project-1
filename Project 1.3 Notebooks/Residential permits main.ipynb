## Correlation between permits issued and value of the building work

```python
value = WA_quarterly_df.iloc[:, 3:6] #construction value
permits = WA_quarterly_df.iloc[:, 0:3] #permits issued

x_values = value
y_values = permits
(slope, intercept, rvalue, pvalue, stderr) = stats.linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel('Construction value for new residential building work')
plt.ylabel('Total permits for new residential building work')
plt.show()
```
