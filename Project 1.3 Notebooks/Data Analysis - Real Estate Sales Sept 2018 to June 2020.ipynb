```python
#Import Dependencies

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import datetime
from scipy.stats import linregress

#set matplotlib params

plt.rcParams['font.size'] = 12
```

```python
#Import CSV into DataFrame

realestate_sales_df = pd.read_csv("aus-property-sales-sep2018-april2020.csv")
realestate_sales_df
```

```python
#Change string value for date_sold to datetimeformat and add quarter, month and year columns for easy sorting.

realestate_sales_df['date_sold'] = pd.to_datetime(realestate_sales_df['date_sold'], format="%Y-%m-%d %H:%M:%S")
realestate_sales_df['quarter'] = realestate_sales_df['date_sold'].dt.quarter
realestate_sales_df['month']= realestate_sales_df['date_sold'].dt.month
realestate_sales_df['year']= realestate_sales_df['date_sold'].dt.year
```

# National real estate sales by state and total volume

```python
#Create national real estate volume data frame for use by Daniel

#Start with a copy for processing

nat_realestate_df = realestate_sales_df

#Group by year and quarter then count another column for total volume. It doesnt matter which column as each entry is a purchase.

nat_realestate_qt_ct_df = nat_realestate_df.groupby(['year','quarter'])['date_sold'].count().reset_index()

#Remove data outside the last quarter of 2018 and the 2nd quarter of 2020

nat_realestate_qt_ct_df = nat_realestate_qt_ct_df[1:-1]
nat_realestate_qt_ct_df.columns = ["Year","Quarter","Sales"]
nat_realestate_qt_ct_df.to_csv("output/national_sales_by_quarter.csv", encoding='utf-8', index=False)

#Create another frame that groups by Year, Quarter, State & Sales.

nat_realestate_qt_ct_st_df = nat_realestate_df.groupby(['year','month','state'])['date_sold'].count().reset_index()
nat_realestate_qt_ct_st_df.columns = ["Year","Quarter","State","Sales"]
nat_realestate_qt_ct_st_df.to_csv("output/state_sales_by_quarter.csv", encoding='utf-8', index=False)
```

```python
#The Same as the previous cell but this time for months.

nat_realestate_mn_ct_df = nat_realestate_df.groupby(['year','month'])['date_sold'].count().reset_index()
nat_realestate_mn_ct_df = nat_realestate_mn_ct_df[1:-1]
nat_realestate_mn_ct_df.columns = ["Year","Month","Sales"]
nat_realestate_mn_ct_df.to_csv("output/national_sales_by_month.csv", encoding='utf-8', index=False)

nat_realestate_mn_ct_st_df = nat_realestate_df.groupby(['year','month','state'])['date_sold'].count().reset_index()
nat_realestate_mn_ct_st_df.columns = ["Year","Month","State","Sales"]
nat_realestate_mn_ct_st_df.to_csv("output/state_sales_by_month.csv", encoding='utf-8', index=False)
```

# Percentage Changes Between Quarter 2 2019 and Quarter 2 2020

```python
#Create a dataframe for sales volume by quarter 2019

wa_realestate_qt_sb_type_2019_2_df
wa_realestate_qt_sb_type_2019_2_df.columns = ["Year","Month","Suburb","Sales_Q22019"]
wa_realestate_qt_sb_type_2019_2_df = wa_realestate_qt_sb_type_2019_2_df.drop(columns=['Year', 'Month'])
wa_realestate_qt_sb_type_2019_2_df
```

```python
#Create a dataframe for sales volume by quarter 2018

wa_realestate_qt_sb_type_2020_2_df
wa_realestate_qt_sb_type_2020_2_df.columns = ["Year","Month","Suburb","Sales_Q22020"]
wa_realestate_qt_sb_type_2020_2_df = wa_realestate_qt_sb_type_2020_2_df.drop(columns=['Year', 'Month'])
wa_realestate_qt_sb_type_2020_2_df
```

```python
#Merge Dataframes By Suburb Dropping those without a match

wa_re_qt_sb_to_20_2_mg = pd.merge(wa_realestate_qt_sb_type_2019_2_df, wa_realestate_qt_sb_type_2020_2_df)
wa_re_qt_sb_to_20_2_mg
```

```python
#Define Percentage Calculation Function

def percCalc(x,y):
    if x<1e-9:
        return 0
    else:
        return (y - x)*100/x
```

```python
#Calculate Percentage Changes Between Q2 2019 and Q2 2020

#Not using. Was supposed to be used with a heatmap. We have far too much to cover in a short period of time and this was more of a nice to have.

wa_re_qt_sb_to_20_2_mg['%change'] = wa_re_qt_sb_to_20_2_mg[['Sales_Q22019','Sales_Q22020']].apply(lambda x: percCalc(x['Sales_Q22019'], x['Sales_Q22020']), axis=1)
wa_re_qt_sb_to_20_2_mg = wa_re_qt_sb_to_20_2_mg.round(2)
wa_re_qt_sb_to_20_2_mg = wa_re_qt_sb_to_20_2_mg.nlargest(50, ['%change'])
wa_re_qt_sb_to_20_2_mg
```
