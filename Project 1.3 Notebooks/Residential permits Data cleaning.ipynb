# Reason for using this data

This documentation was chosen due to the accuracy of the information. Every local government, and on the East coast private issuers of permits, provides statistical information each month to the government. They have been doing this since 1970. even though there is a possibility of errors, they will be minor considering the amount of information gathered and the prolonged period of time this has been happening.

The files have been downloaded per State with the information from 1970. The files have then been adjusted to show the information relevant to answer our question of impact of Covid 19 on the industry.

# Loading all files into the workbook

All files show the top of the table to make sure all the information is the same and accounted for.

All data relates to residential building work separated int he following types:
- Dwellings
- Townhouses: Semi-detached, row, terrace and townhouses all single and multi storey buildings
- Apartments: Flats, units and apartments all single and multi storey buildings
- Other: Additions and alterations

## ACT

```python
ACT_df = pd.read_csv('/Users/sylviabroadbent/Dropbox/Sylvia/UWA_Data_Analytics_Bootcamp/HOMEWORK/7 Project 1/Data-Analytics-Project-1/Project 1.1 Raw Data/Residential construction -  new build, additions/Dwelling units approved, number and value, ACT.csv')
ACT_df.head()
ACT_df
```

```python
ACT_df = ACT_df.iloc[654:675, :]
ACT_df = ACT_df.iloc[:, [0, 1, 4, 8, 11, 14, 18]]
ACT_df
```

```python
ACT_df = ACT_df.rename(columns={
    'Unnamed: 0': '',
    'Total number of dwelling units ;  Australian Capital Territory ;  New ;  Houses ;': 'ACT_dwellings',
    'Total number of dwelling units ;  Australian Capital Territory ;  New ;  Semi-detached, row or terrace houses, townhouses - Total ;': 'ACT_townhouses',
    'Total number of dwelling units ;  Australian Capital Territory ;  New ;  Flats units or apartments - Total including those attached to a house ;': 'ACT_apartments',
    'Total value of building jobs ;  Australian Capital Territory ;  New ;  Houses ;': "ACT_dwellings_$'000",
    'Total value of building jobs ;  Australian Capital Territory ;  New ;  Semi-detached, row or terrace houses, townhouses - Total ;': "ACT_townhouses_$'000",
    'Total value of building jobs ;  Australian Capital Territory ;  New ;  Flats units or apartments - Total including those attached to a house ;': "ACT_apartments_$'000",})
cols = ACT_df.columns.drop('')
ACT_df[cols] = ACT_df[cols].apply(pd.to_numeric)
ACT_df
```

```python
ACT_df[''] = pd.to_datetime(ACT_df[''], format='%b-%Y')
ACT_df = ACT_df.groupby(pd.PeriodIndex(ACT_df[''], freq='Q')).sum()
ACT_df
```

```python
ACT_df['total_permits'] = ACT_df.iloc[:, 0:3].sum(axis=1)
ACT_df['total_value'] = ACT_df.iloc[:, 3:6].sum(axis=1)
ACT_df = ACT_df.iloc[:, 6:8]
ACT_df
```
